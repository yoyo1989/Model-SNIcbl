This notebook
- compare synthetic spectra (cav4 output) to average SN-GRB spectra without and with continuum
- compare sythetic LCs (cav4_eq.lc and cav4_pol.lc) to those of SN Ic-bl shown in figure 8 of [Prentice et al. (2016)](http://mnras.oxfordjournals.org/content/458/3/2973.full.pdf) 


```python
import numpy as np
from scipy.io.idl import readsav
import pylab as pl
import pandas as pd
import readcoll 
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import matplotlib.pyplot as plt
from IPython.display import Image
from IPython.display import display
%matplotlib inline
```

# I. Plots below compare synthetic spectra to average SN Ic-bl spectra without and with continuum

## 1) Compare spectra without continuum

```python
def plotmont1(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        #print spec+phase[i]+'_flat.txt'
        w,f = readcoll.readcol(spec+phase[i]+'_flat.txt',twod=False)
        #print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        #s.fmean=s.fmean*3
        #s.fsdev=s.fsdev*3
        pl.text(4100,shift[i]+1.5,phasemean[i]+'+/-2d',fontsize=20)
        if i == 0:
            pl.plot(w, f*5+shift[i], label=spec, color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f*5+shift[i], color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3)    
        pl.fill_between(s.wlog, s.fmean*5-s.fsdev*5+shift[i], s.fmean*5+s.fsdev*5+shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,40))
plt.subplot(321)
plotmont1('cav4/cav4_eq_', 
         ['12', '17', '22', '27', '32', '37', '42'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_withGRB_1specperSN_', 
         ['-05','00', '05', '10', '15', '20', '25'], 
         [5, 10, 15, 20, 25, 30, 35, 40])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 9500)
pl.ylim(3, 43)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened-equatorial vs average SN-GRB',fontsize=20)


plt.subplot(322)
plotmont1('cav4/cav4_pol_', 
         ['12', '17', '22', '27', '32', '37', '42'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_withGRB_1specperSN_', 
         ['-05','00', '05', '10', '15', '20', '25'], 
         [5, 10, 15, 20, 25, 30, 35, 40])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 9500)
pl.ylim(3, 43)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened-polar vs average SN-GRB',fontsize=20)
```

```python
def plotmont1(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        #print spec+phase[i]+'_flat.txt'
        w,f = readcoll.readcol(spec+phase[i]+'_flat.txt',twod=False)
        #print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        s.fmean=s.fmean*3
        s.fsdev=s.fsdev*3
        pl.text(4100,shift[i]+1.5,phasemean[i]+'+/-2d',fontsize=20)
        if i == 0:
            pl.plot(w, f*5+shift[i], label=spec, color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f*5+shift[i], color='black', linewidth=1)
            pl.plot(s.wlog, s.fmean*5+shift[i], color='red',linewidth=3)    
        pl.fill_between(s.wlog, s.fmean*5-s.fsdev*5+shift[i], s.fmean*5+s.fsdev*5+shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,40))
plt.subplot(321)
plotmont1('cav4/cav4_eq_', 
         ['12', '17', '22', '27', '32', '37', '42'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_withGRB_1specperSN_', 
         ['-05','00', '05', '10', '15', '20', '25'], 
         [5, 10, 15, 20, 25, 30, 35, 40])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 9500)
pl.ylim(3, 43)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened-equatorial vs average SN-GRB',fontsize=20)


plt.subplot(322)
plotmont1('cav4/cav4_pol_', 
         ['12', '17', '22', '27', '32', '37', '42'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_withGRB_1specperSN_', 
         ['-05','00', '05', '10', '15', '20', '25'], 
         [5, 10, 15, 20, 25, 30, 35, 40])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 9500)
pl.ylim(3, 43)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('flattened-polar vs average SN-GRB',fontsize=20)
```

## 2) Compare spectra with continuum

```python
def plotmont2(spec, phase, specmean, phasemean, shift):
    for i in range(len(phase)):
        #print spec+phase[i]+'.txt'
        w,f = readcoll.readcol(spec+phase[i]+'.txt',twod=False)
        f=f/np.mean(f)/2+0.5
        #print specmean+phasemean[i]+'_ft.sav'
        s=readsav(specmean+phasemean[i]+'_ft.sav')
        dwbin=s.wlog[1:1024]-s.wlog[0:1023] # very important
        dwbin=np.append(dwbin[0], dwbin)/20
        pl.text(5200,shift[i]-0.5,phasemean[i]+'+/-2d',fontsize=20)
        
        if i == 0:
            pl.plot(w, f+shift[i], label=spec, color='black')
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3, label='average IcBL')
        else:
            pl.plot(w, f+shift[i], color='black')
            pl.plot(s.wlog, (s.fmean+1)*s.smean/dwbin+shift[i], color='red',linewidth=3)   
        pl.fill_between(s.wlog, (s.fmean-s.fsdev+1)*s.smean/dwbin+shift[i], (s.fmean+s.fsdev+1)*s.smean/dwbin
                        +shift[i], color='red', alpha=0.5)
```

```python
plt.figure(figsize=(20,40))
plt.subplot(321)
plotmont2('cav4/cav4_eq_', 
         ['12', '17', '22', '27', '32', '37', '42'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_withGRB_1specperSN_', 
         ['-05','00', '05', '10', '15', '20', '25'], 
         [5, 10, 15, 20, 25, 30, 35, 40])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(3, 43)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('equatorial vs average SN-GRB',fontsize=20)


plt.subplot(322)
plotmont2('cav4/cav4_pol_', 
         ['12', '17', '22', '27', '32', '37', '42'], 
         'OneSpectrumPerSNIcbl/meanspecIcBL_withGRB_1specperSN_', 
         ['-05','00', '05', '10', '15', '20', '25'], 
         [5, 10, 15, 20, 25, 30, 35, 40])
pl.xlabel("Rest wavelength ($\AA$)",fontsize=20)
pl.ylabel("Relative flux + constant",fontsize=20)
pl.xlim(3500, 8500)
pl.ylim(3, 43)
pl.legend(fontsize=15)
plt.tick_params(labelsize=20)
pl.title('polar vs average SN-GRB',fontsize=20)
```

# II. Light curves

The shape of synthetic LCs are very similar to LCs of SNe Ic-bl sample in [Prentice et al. (2016)](http://mnras.oxfordjournals.org/content/458/3/2973.full.pdf). In particular, the rise time (half luminosity to peak luminosity) and the decay time (peak luminosity to half luminosity) of synthetic LCs are consistent with those of observed SNe Ic-bl in Prentice et al. (2016), which are 8-12 days and 13-16 days from their table 6, respectively.

```python
plt.figure(figsize=(8,4))
t,L = readcoll.readcol('cav4/cav4_eq.lc',twod=False)
pl.plot(t-17.5, np.log10(L), label='equatorial', color='black')
pl.plot([-20,40],[np.log10(np.max(L)/2),np.log10(np.max(L)/2)],color='black',
        ls='--',label='half luminosity')
t,L = readcoll.readcol('cav4/cav4_pol.lc',twod=False)
pl.plot(t-17.5, np.log10(L), label='polar', color='blue')
pl.plot([-20,40],[np.log10(np.max(L)/2),np.log10(np.max(L)/2)],color='blue',
        ls='--',label='half luminosity')
pl.xlabel("Days WRT bolometric max",fontsize=20)
pl.ylabel("Log(Luminosity)",fontsize=20)
pl.xlim(-19, 35)
pl.ylim(41.4, 43)
pl.legend(fontsize=15,loc='lower right', ncol=2)
pl.title('Synthetic Bolometric LC',fontsize=20)
```
